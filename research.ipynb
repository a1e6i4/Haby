```python
import pymorphy2

# Создание экземпляра класса MorphAnalyzer
morph = pymorphy2.MorphAnalyzer()

# Фраза с несогласующимися падежами
phrase = "Цель создана: 10000 фразы к 23.11.2023 Чтоб успеть к сроку вам надо ругаться матом -243.9 фразы в день"

# Разбиваем фразу на слова
words = phrase.split()

# Исправляем падежи
corrected_phrase = []
for word in words:
    parsed_word = morph.parse(word)[0]  # Анализируем каждое слово
    normalized_word = parsed_word.inflect({'nomn'}).word  # Исправляем на именительный падеж
    corrected_phrase.append(normalized_word)

# Объединяем слова обратно в предложение
corrected_sentence = ' '.join(corrected_phrase)

print("Исправленная фраза:", corrected_sentence)
```
